```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
size = 50
iterations = 50
l_rad = 1
r_rad = 1
diameter = l_rad + r_rad + 1
# rule = np.random.randint(low=0, high=2**(2**diameter), dtype='uint64')
rule = 184

state = [0, 1] * (size//2)
np.random.shuffle(state)

# state = np.zeros(size)
# state[size//2] = 1

rule_bin = format(rule, f'0{2**diameter}b')
evolution = np.ndarray(shape=(iterations, size))
evolution[0] = state

lut = {}
for i in range(2**diameter):
    lut[format(2**diameter-i-1, f'0{diameter}b')] = rule_bin[i]

for i, state in enumerate(evolution[:-1]):
    for j in range(size):
        config = ''.join([str(int(state[k % size])) for k in range(j-l_rad, j+r_rad+1)])
        evolution[i+1, j] = lut[config]
```

```python
plt.figure(figsize=(20,20))
plt.imshow(evolution, cmap='Greys', interpolation='nearest')
# plt.show()
# plt.savefig(str(datetime.now()).replace(':', '-') + f'_{rule}.png')
```
